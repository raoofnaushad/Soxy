```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=7008,
                                                    train_size=0.8, shuffle=True)

print(f"Size of training data: {X_train.shape} and count of test data: {X_test.shape}")
print(f"Size of training labels: {y_train.shape} and count of test labels: {y_test.shape}")
```

## Model - 1
**Model 1 is a model with no hidden layer. 4 neuron input layer and 1 neuron output layer**

### More about the model.
- We are using fully connected input layer to an output layer to predict for a binary classification problem.
- **Sigmoid function** is used as the final activation layer. This function will bring in non-linearity to the neural network. This will scale all the outputs between [0, 1]
- **Binary Cross Entropy loss** is used as it is one of the best loss function for Binary Classification. it encourages the model to output high probabilities for the correct class (1) and low probabilities for the incorrect class (0). It penalizes deviations from the true binary labels effectively, making it a suitable choice for optimizing binary classification tasks.
- **L2 regularization** is used to avoid overfitting. L2 regularization helps prevent overfitting by adding a penalty term to the loss function during training. This penalty term discourages the model from assigning excessively large weights to any particular feature.
- **Early Stopping** is also used to avoid overfitting. This will stop training when there is no further improvement in validation accuracy.

### Result Evaluation.
- We got a training accuracy of 98.72% and a validation accuracy of 99.27%. The model training stopped at 8th Epoch.
- **From the graph**, we can understand that the training accuracy is increasing gradually but the validation accuracy has an exponential increase intially and then at the 4th epoch it reached its maximum.

### Logistic Regression and Model 1 are same theoretically and practically too. Why there is a difference?
- **Compared to Logistic Regression** we have less training accuracy. But the validation accuracy was same. The difference in training accuracy can be because the training has stopped immediately after there is no further improvement in validation accuracy which matters most. Also, the random weight initialization methods in sklearn might be more efficient than what we used here.

```python
net = N.Network() ## Creating network object
net.add(L.FCLayer(4, 1)) ## Adding classifier Layer with input size 4 and output size 1.
net.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad)) ## Adding sigmoid activation layer 
net.assign(F.BCELoss, F.BCELoss_grad) ## Assigning Binary Cross Entropy Loss as Loss Function
```

```python
## Defining hyper parameters based on the hyper parameter tuning did at the end of this notebook
lr = 0.01
lmbda = 0.001
epochs = 100
train_acc, val_acc, epochs_taken = net.train(X_train, 
                                           y_train, 
                                           X_test, 
                                           y_test, 
                                           learning_rate=lr, 
                                           epochs=epochs, 
                                           lmda=lmbda, 
                                           early_stopping=True, 
                                           print_metrics=False)

print(f"Training Accuracy: {train_acc}, Validation Accuracy: {val_acc}, and the epochs taken: {epochs_taken}")
```

```python
U.plot_accuracy_graph(net.train_accuracies, net.val_accuracies, "Model 1")
```

## Hyper Parameter Tuning with cross validation - Lambda and Learning Rate

### Advantage of using k-fold cross validation for hyper parameter tuning
- K-fold cross-validation is a technique that partitions your dataset into 'k' equally sized folds, allowing you to train and evaluate your model 'k' times.
- In each iteration, one of the 'k' folds serves as a validation set while the remaining 'k-1' folds are used for training.
- This process provides a more robust assessment of your model's performance, aids in hyperparameter tuning by reducing overfitting, and ensures that the model generalizes well to unseen data

### Evaluation of Cross Validation
- Best Validation Accuracy obtained is 98.982%. This gives us the right understanding how well our model works in real world. Since it is the average of accuracies after evaluating multiple sets of data.
- Best Lambda is identified as 0.001
- Best learning rate is identified as 0.1

```python
X = np.expand_dims(X.to_numpy().astype(np.float32), axis=1)
y = np.expand_dims(y.to_numpy().astype(np.float32), axis=1)


lr, epochs, lmbda, early_stopping, print_metrics = 0.01, 100, None, True, False

## Lambda Hyper Parameter Tuning...
lambdas = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1.0]
## Learning rate Hyper Parameter Tuning...
learning_rates = [0.001, 0.01, 0.1, 1.0]

# Initialize variables to store best hyperparameters and accuracy
best_lambda = None
best_learning_rate = None
best_accuracy = 0.0

# Perform cross-validation with hyperparameter tuning
print("Performing tuning for lambda hyper parameter")
print("-----"*10)
for lmbda in lambdas:
    for lr in learning_rates:
        ## Creating model
        cross_net = N.Network()
        cross_net.add(L.FCLayer(4, 1))
        cross_net.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad))
        cross_net.assign(F.BCELoss, F.BCELoss_grad)
        ## Doing K-Fold Cross Validation
        avg_val_accuracy, lr, lmbda = U.kfold_cross_validation(X, y, cross_net, lr, epochs, lmbda, early_stopping, print_metrics, num_folds=5)
        if avg_val_accuracy > best_accuracy:
            best_accuracy = avg_val_accuracy
            best_lambda = lmbda
            best_learning_rate = lr
        print(f"Average accuracy with lr={lr}, lambda={lmbda} is: {round(avg_val_accuracy, 3)}")
        
        
print(f"Best Lambda: {best_lambda}")
print(f"Best learning rate: {best_learning_rate}")
print(f"Best Validation Accuracy: {best_accuracy}")
```
