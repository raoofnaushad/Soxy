# Neural Network Modelling

## Processing data for NN

````python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=7008,
                                                    train_size=0.8, shuffle=True)

print(f"Size of training data: {X_train.shape} and count of test data: {X_test.shape}")
print(f"Size of training labels: {y_train.shape} and count of test labels: {y_test.shape}")
````

````python
X_train = np.expand_dims(X_train.to_numpy().astype(np.float32), axis=1)
y_train = np.expand_dims(y_train.to_numpy().astype(np.float32), axis=1)
X_test = np.expand_dims(X_test.to_numpy().astype(np.float32), axis=1)
y_test = y_test.to_numpy().astype(np.float32)
print(f"Size of training data: {X_train.shape} and count of test data: {X_test.shape}")
print(f"Size of training labels: {y_train.shape} and count of test labels: {y_test.shape}")
````

## Model Creating with no hidden layers with best params after hyper parameter tuning

````python
net = N.Network() ## Creating network object
net.add(L.FCLayer(4, 1)) ## Adding classifier Layer with input size 4 and output size 1.
net.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad)) ## Adding sigmoid activation layer 
net.assign(F.BCELoss, F.BCELoss_grad) ## Assigning Binary Cross Entropy Loss as Loss Function
````

````python
## Defining hyper parameters based on the hyper parameter tuning did at the end of this notebook
lr = 0.01
lmbda = 0.001
epochs = 100
train_acc, val_acc, best_epoch = net.train(X_train, 
                                           y_train, 
                                           X_test, 
                                           y_test, 
                                           learning_rate=lr, 
                                           epochs=epochs, 
                                           lmda=lmbda, 
                                           early_stopping=True, 
                                           print_metrics=False)
````

````python
U.plot_accuracy_graph(net.train_accuracies, net.val_accuracies, "Model 1")
````

## Model Creating with 1 hidden layers (3 Neurons) with best params after hyper parameter tuning

````python
net2 = N.Network()
net2.add(L.FCLayer(4, 3))
net2.add(L.ActivationLayer(F.tanh, F.tanh_grad))
net2.add(L.FCLayer(3, 1))
net2.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad))
net2.assign(F.BCELoss, F.BCELoss_grad)
````

````python
## Defining hyper parameters based on the hyper parameter tuning did at the end of this notebook
lr = 0.01
lmbda = 0.001
epochs = 100
train_acc, val_acc, best_epoch = net2.train(X_train, y_train, X_test, y_test, learning_rate=lr, epochs=epochs, lmda=lmbda, early_stopping=True, print_metrics=False)
# net.train(X_train, y_train, X_test, y_test, learning_rate=lr, epochs=epochs, lmda=lmbda, early_stopping=True, print_metrics=True)
````

## Comparison on Regularization

````python
## training without L2 regularization

net_noreg = N.Network() ## Creating network object
net_noreg.add(L.FCLayer(4, 1)) ## Adding classifier Layer with input size 4 and output size 1.
net_noreg.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad)) ## Adding sigmoid activation layer 
net_noreg.assign(F.BCELoss, F.BCELoss_grad) ## Assigning Binary Cross Entropy Loss as Loss Function

## Defining hyper parameters based on the hyper parameter tuning did at the end of this notebook
lr = 0.01
lmbda = None
epochs = 100
train_acc, val_acc, best_epoch = net_noreg.train(X_train, y_train, X_test, y_test, learning_rate=lr, epochs=epochs, lmda=lmbda, early_stopping=True, print_metrics=False)
````

## Hyper Parameter Tuning - Lambda and Learning Rate

````python
## Lambda Hyper Parameter Tuning...
lambdas = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1.0]
## Learning rate Hyper Parameter Tuning...
learning_rates = [0.001, 0.01, 0.1, 1.0]

# Initialize variables to store best hyperparameters and accuracy
best_lambda = None
best_learning_rate = None
best_accuracy = 0.0

# Perform cross-validation with hyperparameter tuning
print("Performing tuning for hyperparameters - Lambda and Learning Rates.")
for lmbda in lambdas:
    for lr in learning_rates:
        ## Change the model two checkout in different models
        net = N.Network()
        net.add(L.FCLayer(4, 1))
        net.add(L.ActivationLayer(F.sigmoid, F.sigmoid_grad))
        net.assign(F.BCELoss, F.BCELoss_grad)
        # val_acc = net.train(X_train, y_train, X_test, y_test, learning_rate=0.01, epochs=100, lmda=lmbda, print_metrics=False) ## With L2 Regularization
        _, val_acc, epoch = net.train(X_train, y_train, X_test, y_test, learning_rate=lr, epochs=100, lmda=lmbda, early_stopping=True, print_metrics=False) ## With L2 and Early Stopping
        if val_acc > best_accuracy:
            best_accuracy = val_acc
            best_lambda = lmbda
            best_learning_rate = lr

print(f"------"*20)
print(f"Best Lambda: {best_lambda}")
print(f"Best learning rate: {best_learning_rate}")
print(f"Best Vaidation Accuracy: {best_accuracy}")
print(f"Best Number of Epochs: {epoch}")
````
